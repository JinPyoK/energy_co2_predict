```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load Data
df = pd.read_csv('South Korea_scaled.csv')

# Set the 'Year' column as the independent variable and the 'Average CO2 emission factor (tCO2/toe)' column as the dependent variable
X = df['Year'].values.reshape(-1, 1)
y = df['Average CO2 emission factor (tCO2/toe)'].values

#  Create Multiple Regression Model
model = LinearRegression()

#  Model Learning
model.fit(X, y)

#  Compare the actual value with the predicted value
y_pred = model.predict(X)

#  Calculate evaluation indicators
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')

#  Calculation of regression line equations
slope = model.coef_[0]
intercept = model.intercept_
equation = f'y = {slope:.2f}x + {intercept:.2f}'

#  Draw scatterplots and regression lines
plt.scatter(X, y, alpha=0.7)
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Average CO2 emission factor (tCO2/toe)')
plt.title('CO2 Emissions and Year in South Korea')
plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.show()
#  Draw scatterplots and regression lines
plt.scatter(y, y_pred, alpha=0.7)
plt.plot([y.min(),y_pred.max()],[y.min(),y.max()],color='red',linestyle='--',linewidth=2,label='Regression Line')
plt.xlabel('Actual value')
plt.ylabel('Predicted value')
plt.title('Multiple Linear Regression-South Korea')
plt.legend()
plt.show()
```
